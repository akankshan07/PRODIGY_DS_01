**BAR PLOT**

```python
Country = ['India', 'Afghanistan', 'Germany', 'Switzerland', 'United States']
population = [7494580, 40099460, 2341185, 612989, 9365149]

# Create a bar plot
plt.figure(figsize=(5, 4))
plt.bar(Country, population, color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Total Population By Country')

plt.xticks(rotation=45)
# Display the plot
plt.tight_layout()
plt.show()
```

**AGE DISTRIBUTION**

```python
Age = ['0-4 years','5-14 years','15-24 years','25-64 years','65+']
Country = ['India', 'Afghanistan', 'Germany', 'Switzerland', 'United States']

# Create a bar chart
plt.figure(figsize=(5, 3))
plt.bar(Country, Age, color='orange')
plt.xlabel('Country')
plt.ylabel('Age')
plt.title('Age Distribution')
plt.xticks(rotation=45)
# Display the plot
plt.tight_layout()
plt.show()
```

```python
Age = ['0-4 years','5-14 years','15-24 years','25-64 years','65+']
Country = ['India', 'Afghanistan', 'Germany', 'Switzerland', 'United States']
population = [7494580, 40099460, 2341185, 612989, 9365149]

# Create a bar-line chart
fig, ax1 = plt.subplots(figsize=(7, 4))

# Create the bar chart on the primary y-axis (left)
ax1.bar(Country, population , color='skyblue')
ax1.set_xlabel('Country')
ax1.set_ylabel('Total Population', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Create the line chart on the secondary y-axis (right)
ax2 = ax1.twinx()
ax2.plot(Country, Age , marker='o', color='orange', linestyle='-', linewidth=2)
ax2.set_ylabel('Average Age', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title('Age Distribution in Total Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
